# Algoritmos de Ordenação

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def heapify(arr, n, i):
    largest = i  # Inicializa o maior como o raiz
    left_child = 2 * i + 1
    right_child = 2 * i + 2

    # Verifica se o filho da esquerda existe e é maior que a raiz
    if left_child < n and arr[left_child] > arr[largest]:
        largest = left_child

    # Verifica se o filho da direita existe e é maior que a raiz
    if right_child < n and arr[right_child] > arr[largest]:
        largest = right_child

    # Se o maior não for a raiz, troca-os
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        # Heapify recursivamente a subárvore afetada
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    # Constrói um max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extrai os elementos, um por um
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Troca
        heapify(arr, i, 0)
```

```python
import random

#Gerador de numeros de pares aleatórios
def generate_random_pairs(num_pairs, min_val=1, max_val=100):
    return [(random.randint(min_val, max_val), random.randint(min_val, max_val)) for _ in range(num_pairs)]

#Calcula a razão R para um conjunto de pares
def calculate_R(pairs):
    sum_x = sum(pair[0] for pair in pairs)
    sum_y = sum(pair[1] for pair in pairs)
    return sum_x / sum_y

#Utiliza Bubble Sort para ordenar os pares em ordem descrescente com base em R e seleciona os que satisfazem o lema (at/bt > R*)
def bubble_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    #Ordena Pares
    for i in range(n):
        for j in range(0, n - i - 1):
            if pairs[j][0] / pairs[j][1] > R:
                pairs[j], pairs[j + 1] = pairs[j + 1], pairs[j]
    #Adiciona pares com razão maior que R, adiciona nos pares selecionados e recalcula R.
    for i in range(n):
        if pairs[i][0] / pairs[i][1] > R:
            selected_pairs.append(pairs[i])
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Gerar pares aleatórios
required_pair = generate_random_pairs(1)[0]
random_pairs = generate_random_pairs(100)

# Resolver o Problema de Programação Hiperbólica com Bubble Sort
selected_pairs, max_R = bubble_sort_with_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

###PPH com ordenação por Insertion Sort

```python
#Utiliza Insertion Sort para ordenar os pares em ordem descrescente com base em R e seleciona os que satisfazem o lema (at/bt > R*)
def insertion_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    for i in range(1, n):
        key = pairs[i]
        j = i - 1
        while j >= 0 and key[0] / key[1] > R:
            pairs[j + 1] = pairs[j]
            j -= 1
        pairs[j + 1] = key

    for i in range(n):
        if pairs[i][0] / pairs[i][1] > R:
            selected_pairs.append(pairs[i])
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Resolver o Problema de Programação Hiperbólica com Insertion Sort
selected_pairs, max_R = insertion_sort_with_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

###PPH com ordenação por Selection Sort

```python
def selection_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if pairs[j][0] / pairs[j][1] > pairs[min_idx][0] / pairs[min_idx][1]:
                min_idx = j
        pairs[i], pairs[min_idx] = pairs[min_idx], pairs[i]

    for i in range(n):
        if pairs[i][0] / pairs[i][1] > R:
            selected_pairs.append(pairs[i])
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Resolver o Problema de Programação Hiperbólica com Selection Sort
selected_pairs, max_R = selection_sort_with_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

## Item II) - Algoritimos O(Nlogn)

### Merge Sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i][0] / left[i][1] > right[j][0] / right[j][1]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

def merge_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    sorted_pairs = merge_sort(pairs)
    selected_pairs = [required_pair]

    for pair in sorted_pairs:
        if pair[0] / pair[1] > R:
            selected_pairs.append(pair)
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Resolver o Problema de Programação Hiperbólica com Merge Sort
selected_pairs, max_R = merge_sort_with_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

###Quick-Sort

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j][0] / arr[j][1] > pivot[0] / pivot[1]:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort_with_pph(arr, low, high, R):
    if low < high:
        pi = partition(arr, low, high)

        # Divida e conquiste
        if arr[pi][0] / arr[pi][1] > R:
            quick_sort_with_pph(arr, low, pi - 1, R)
            quick_sort_with_pph(arr, pi + 1, high, R)

def quick_sort_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    quick_sort_with_pph(pairs, 0, n - 1, R)

    for pair in pairs:
        if pair[0] / pair[1] > R:
            selected_pairs.append(pair)
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Resolver o Problema de Programação Hiperbólica com Quick-Sort
selected_pairs, max_R = quick_sort_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

###Heap-sort


```python
import heapq
def heap_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    # Construir um heap mínimo (min-heap) com base na razão ai/bi
    min_heap = [(pair[0] / pair[1], pair) for pair in pairs]
    heapq.heapify(min_heap)

    while min_heap:
        ratio, pair = heapq.heappop(min_heap)
        if pair[0] / pair[1] > R:
            selected_pairs.append(pair)
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Resolver o Problema de Programação Hiperbólica com Heap-Sort
selected_pairs, max_R = heap_sort_with_pph(random_pairs, required_pair)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

```python
import random

def calculate_ratio(pair):
    return pair[0] / pair[1]

def median_of_medians(pairs, k):
    if len(pairs) <= 5:
        return sorted(pairs, key=calculate_ratio)[k]

    groups = [pairs[i:i+5] for i in range(0, len(pairs), 5)]
    medians = [median_of_medians(group, len(group)//2) for group in groups]

    pivot = median_of_medians(medians, len(medians)//2)

    lower = [pair for pair in pairs if calculate_ratio(pair) < calculate_ratio(pivot)]
    upper = [pair for pair in pairs if calculate_ratio(pair) > calculate_ratio(pivot)]

    if k < len(lower):
        return median_of_medians(lower, k)
    elif k < len(pairs) - len(upper):
        return pivot
    else:
        return median_of_medians(upper, k - (len(pairs) - len(upper)))

def quick_sort_median_of_medians(arr):
    if len(arr) <= 1:
        return arr

    pivot = median_of_medians(arr, len(arr) // 2)
    lower = [pair for pair in arr if calculate_ratio(pair) < calculate_ratio(pivot)]
    equal = [pair for pair in arr if calculate_ratio(pair) == calculate_ratio(pivot)]
    upper = [pair for pair in arr if calculate_ratio(pair) > calculate_ratio(pivot)]

    return quick_sort_median_of_medians(lower) + equal + quick_sort_median_of_medians(upper)

# Gerar pares aleatórios
random_pairs = [(random.randint(1, 100), random.randint(1, 100)) for _ in range(100)]

# Ordenar os pares usando o Quick-Sort com Mediana das Medianas
sorted_pairs = quick_sort_median_of_medians(random_pairs)

print(f"Sorted Pairs: {sorted_pairs}")
```

```python
def calculate_R(pairs):
    sum_x = sum(pair[0] for pair in pairs)
    sum_y = sum(pair[1] for pair in pairs)
    return sum_x / sum_y

def hoare_partition(arr, low, high, pivot):
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while True:
            i_ratio = (arr[i][0] + sum(pair[0] for pair in arr[low:i])) / (arr[i][1] + sum(pair[1] for pair in arr[low:i]))
            if i_ratio >= pivot:
                break
            i += 1

        j -= 1
        while True:
            j_ratio = (arr[j][0] + sum(pair[0] for pair in arr[j+1:high+1])) / (arr[j][1] + sum(pair[1] for pair in arr[j+1:high+1]))
            if j_ratio <= pivot:
                break
            j -= 1

        if i >= j:
            return j

        arr[i], arr[j] = arr[j], arr[i]

def quick_sort_pph_hoare(arr, low, high, required_pair):
    if low < high:
        pivot = (required_pair[0] + sum(pair[0] for pair in arr[low:high])) / (required_pair[1] + sum(pair[1] for pair in arr[low:high]))
        pi = hoare_partition(arr, low, high, pivot)

        # Divida e conquiste
        quick_sort_pph_hoare(arr, low, pi, required_pair)
        quick_sort_pph_hoare(arr, pi + 1, high, required_pair)

def solve_PPH_hoare(required_pair, pairs):
    n = len(pairs)

    # Inicialmente, ordena os pares com base na razão ai/bi
    pairs.sort(key=lambda pair: pair[0] / pair[1], reverse=True)

    # Chama o Quick-Sort com Particionamento de Hoare adaptado para PPH
    quick_sort_pph_hoare(pairs, 0, n - 1, required_pair)

    # Seleciona os pares que satisfazem a condição at/bt > R
    R = calculate_R([required_pair])
    selected_pairs = [required_pair]

    for pair in pairs:
        ratio = pair[0] / pair[1]
        if ratio >= R:
            selected_pairs.append(pair)
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Exemplo de uso
required_pair = (10, 20)
pairs = [(5, 10), (15, 30), (3, 6), (7, 14), (8, 16)]

selected_pairs, max_R = solve_PPH_hoare(required_pair, pairs)

print(f"Required Pair: {required_pair}")
print(f"Selected Pairs: {selected_pairs}")
print(f"Max R: {max_R}")
```

```python
# Gerador de números de pares aleatórios
def generate_random_pairs(num_pairs, min_val=1, max_val=10000):
    return [(random.randint(min_val, max_val), random.randint(min_val, max_val)) for _ in range(num_pairs)]

# Função para calcular a razão R
def calculate_R(pairs):
    sum_x = sum(pair[0] for pair in pairs)
    sum_y = sum(pair[1] for pair in pairs)
    return sum_x / sum_y

# Tamanhos de entrada
input_sizes = [100, 500, 1000, 5000, 7500 , 10000,50000, 75000, 100000, 200000]

# Gerar pares aleatórios uma única vez
random_pairs = generate_random_pairs(max(input_sizes))
```

##Questão I: PPH com Medições


```python
# Função para resolver o problema com PPH
def solve_PPH(required_pair, pairs):
    selected_pairs_indices = []
    selected_pairs = [required_pair]
    max_R = calculate_R(selected_pairs)
    history_R = [max_R]

    change_R = True
    while change_R and len(selected_pairs) < len(pairs) + 1:
        change_R = False
        for i, pair in enumerate(pairs):
            if pair[0] / pair[1] > max_R:
                selected_pairs_indices.append(i)
                selected_pairs.append(pair)
                max_R = calculate_R(selected_pairs)
                history_R.append(max_R)
                change_R = True

    return selected_pairs_indices, history_R

# Função para medir tempo de execução e memória
def measure_execution_time_and_memory(pairs, required_pair):
    start_time = time.time()
    selected_indices, history_R = solve_PPH(required_pair, pairs)
    end_time = time.time()

    memory_usage = psutil.Process().memory_info().rss
    execution_time = end_time - start_time

    return execution_time, memory_usage

# Dicionários para armazenar tempos de execução e uso de memória
execution_times = {size: [] for size in input_sizes}
memory_usages = {size: [] for size in input_sizes}

# Loop através dos tamanhos de entrada
for size in input_sizes:
    total_execution_time = 0
    num_executions = 0

    while total_execution_time < 5:
        random_pairs = generate_random_pairs(size)
        required_pair = generate_random_pairs(1)[0]

        execution_time, memory_usage = measure_execution_time_and_memory(random_pairs, required_pair)
        total_execution_time += execution_time
        num_executions += 1

    execution_times[size].append(total_execution_time / num_executions)
    memory_usages[size].append(memory_usage)

# DataFrame para armazenar os resultados
results_pph = []
for size in input_sizes:
    avg_execution_time = execution_times[size][0]  # Como agora estamos considerando a média, pegamos o valor único na lista.
    avg_memory_usage = memory_usages[size][0]  # Similarmente, pegamos o valor único.
    results_pph.append((size, avg_execution_time, avg_memory_usage))

df1 = pd.DataFrame(results_pph, columns=["Tamanho da Entrada", "Tempo Médio (s)", "Uso Médio de Memória (bytes)"])

# Imprimir tabela de resultados
print(df1)

# Gráficos para tempo médio de execução
plt.figure(figsize=(10, 5))
for size in input_sizes:
    plt.plot([size], execution_times[size], 'o', label=f'Tamanho {size}')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Tempo de Execução Médio (s)')
plt.legend()
plt.title('Tempo de Execução Médio')
plt.show()

# Gráficos para uso médio de memória
plt.figure(figsize=(10, 5))
for size in input_sizes:
    plt.plot([size], memory_usages[size], 'o', label=f'Tamanho {size}')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Uso de Memória Médio (bytes)')
plt.legend()
plt.title('Uso de Memória Médio')
plt.show()
```

```python
# Função de ordenação Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i][0] / left[i][1] > right[j][0] / right[j][1]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# Utiliza Merge Sort para ordenar os pares em ordem decrescente com base em R e seleciona os que satisfazem o lema (at/bt > R*)
def merge_sort_with_pph(pairs, required_pair):
    n = len(pairs)
    R = calculate_R([required_pair])
    sorted_pairs = merge_sort(pairs)
    selected_pairs = [required_pair]

    for pair in sorted_pairs:
        if pair[0] / pair[1] > R:
            selected_pairs.append(pair)
            R = calculate_R(selected_pairs)

    return selected_pairs, R

# Função para medir tempo de execução e memória
def measure_execution_time_and_memory(pairs, required_pair):
    start_time = time.time()
    selected_pairs, max_R = merge_sort_with_pph(pairs, required_pair)
    end_time = time.time()

    memory_usage = psutil.Process().memory_info().rss
    execution_time = end_time - start_time

    return execution_time, memory_usage

# Dicionários para armazenar tempos de execução e uso de memória
execution_times = {size: [] for size in input_sizes}
memory_usages = {size: [] for size in input_sizes}

# Loop através dos tamanhos de entrada
for size in input_sizes:
    required_pair = generate_random_pairs(1)[0]

    # Use os mesmos pares aleatórios gerados uma única vez
    random_pairs_for_size = random_pairs[:size]

    total_execution_time = 0
    num_executions = 0

    while total_execution_time < 5:
        execution_time, memory_usage = measure_execution_time_and_memory(random_pairs_for_size, required_pair)
        total_execution_time += execution_time
        num_executions += 1

    execution_times[size].append(total_execution_time / num_executions)
    memory_usages[size].append(memory_usage)

# DataFrame para armazenar os resultados
results_merge = []
for size in input_sizes:
    avg_execution_time = execution_times[size][0]  # Como agora estamos considerando a média, pegamos o valor único na lista.
    avg_memory_usage = memory_usages[size][0]  # Similarmente, pegamos o valor único.
    results_merge.append((size, avg_execution_time, avg_memory_usage))

df5 = pd.DataFrame(results_merge, columns=["Tamanho da Entrada", "Tempo Médio (s)", "Uso Médio de Memória (bytes)"])

# Imprimir tabela de resultados
print(df5)

# Gráficos para tempo médio de execução
plt.figure(figsize=(10, 5))
for size in input_sizes:
    plt.plot([size], execution_times[size], 'o', label=f'Tamanho {size}')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Tempo de Execução Médio (s)')
plt.legend()
plt.title('Tempo de Execução Médio')
plt.show()

# Gráficos para uso médio de memória
plt.figure(figsize=(10, 5))
for size in input_sizes:
    plt.plot([size], memory_usages[size], 'o', label=f'Tamanho {size}')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Uso de Memória Médio (bytes)')
plt.legend()
plt.title('Uso de Memória Médio')
plt.show()
```
