```python
import random
import time
import tracemalloc


def criar_vetores(n, peso, valor):
    # Função para criar vetores de pesos e valores aleatórios
    pesos = [random.randint(1, peso) for _ in range(n)]
    valores = [random.randint(1, valor) for _ in range(n)]
    return pesos, valores


def mochila_fracionaria_nlogn(pesos, valores, capacidade):
    # Cria uma lista de tuplas que armazena a razão valor/peso, pesos e valores, --- facilitar a organização
    razao = [(valores[i] / pesos[i], pesos[i], valores[i]) for i in range(len(pesos))]


    # Ordena a lista de tuplas em ordem decrescente da razão valor/peso ---primeiro elemento melhor valor possivel.
    razao.sort(reverse=True)

    valor_maximo = 0
    mochila = []
    quant_fracionaria = []
    #Preenche a mochila de acordo com a ordem da lista razao, já que os priemiros elementos serão os melhores de se colocar na mochila devido a razão --- complexidade varia o(1) o pior o(n)
    for i in range(len(razao)):
        capacidade_usada = min(1, capacidade / razao[i][1]) * razao[i][1]
        capacidade -= capacidade_usada
        if capacidade_usada > 0:
            frac_item = capacidade_usada / razao[i][1]
            quant_fracionaria.append(frac_item)
            valor_maximo += frac_item * razao[i][2]
            mochila.append([razao[i][1], razao[i][2]])

    return valor_maximo, mochila, quant_fracionaria




def pivot_item_c(pesos, valores, capacidade, inicio, fim):
    # Calcular a mediana das razões valor/peso
    razoes = [pesos[i] / valores[i] for i in range(inicio, fim + 1)]
    pivot = round(sum(razoes) / len(razoes), 0)#calcula o pivot

    menor = []  # Lista para índices dos itens menores que o pivot
    igual = []  # Lista para índices dos itens iguais ao pivot
    maior = []  # Lista para índices dos itens maiores que o pivot

    # Itera pelos índices dos itens no intervalo [inicio, fim]
    for i in range(inicio, fim + 1):
        if valores[i] < pivot:
            menor.append(i)
        elif valores[i] == pivot:
            igual.append(i)
        else:
            maior.append(i)

    return menor, igual, maior  # Retorna as listas de índices


def mochila_fracionaria_com_pivot_item_c(pesos, valores, capacidade, inicio, fim):
    if inicio <= fim:
        # Chama a função de particionamento com pivot
        menor, igual, maior = pivot_item_c(pesos, valores, capacidade, inicio, fim)
        valor_max = 0
        mochila = []
        frac = []

        # Se não houver itens maiores que o pivot, todos os itens iguais e menores são escolhidos
        if not maior:
            for i in igual + menor:
                capacidade_usada = min(capacidade, pesos[i])
                if capacidade_usada > 0:
                    frac_item = capacidade_usada / pesos[i]
                    frac.append(frac_item)
                    valor_max += frac_item * valores[i]
                    mochila.append([pesos[i], valores[i]])
                    capacidade -= capacidade_usada
        else:
            for i in maior + igual + menor:
                capacidade_usada = min(capacidade, pesos[i])
                if capacidade_usada > 0:
                    frac_item = capacidade_usada / pesos[i]
                    frac.append(frac_item)
                    valor_max += frac_item * valores[i]
                    mochila.append([pesos[i], valores[i]])
                    capacidade -= capacidade_usada

    return valor_max, mochila, frac

#calculo do tempo de execução e memória consumida

def calcular_media_tempo_memoria(funcao):
    tempos = []
    memorias = []

    tempos_max = float('-inf')
    tempos_min = float('inf')
    memorias_max = 0
    memorias_min = float('inf')

    for _ in range(5):
        tracemalloc.start()
        start_time = time.time()

        # Chame a função a ser medida aqui
        funcao()

        end_time = time.time()
        current, _ = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        tempo_execucao = end_time - start_time
        tempos.append(tempo_execucao)
        memorias.append(current / 10**6)  # Converta para MB

        # Atualize os valores máximos e mínimos
        tempos_max = max(tempos_max, tempo_execucao)
        tempos_min = min(tempos_min, tempo_execucao)
        memorias_max = max(memorias_max, current / 10**6)
        memorias_min = min(memorias_min, current / 10**6)

    media_tempo = sum(tempos) / len(tempos)
    media_memoria = sum(memorias) / len(memorias)

    return media_tempo, media_memoria, tempos_max, tempos_min, memorias_max, memorias_min





#======================================Implementação==================================================
n = 3000000 # Número de vetores
peso_maximo = 30  # Peso máximo de um item
valor_maximo = 10  # Valor máximo de um item
capacidade = 40 #Considera como capacidade máxima um valor de 40
vetor_pesos,vetor_valores = criar_vetores(n,peso_maximo,valor_maximo)

print("Aplicação da função moclila_fracionaria_On_tiem_c")
media_tempo1, media_memoria1,temp_max, temp_mim,memori_max, memori_min = calcular_media_tempo_memoria(lambda:mochila_fracionaria_com_pivot_item_c(vetor_pesos, vetor_valores, capacidade,0, len(vetor_pesos)-1))
print(f"Tempo máximo de execução: {temp_max} segundos")
print(f"Tempo mínimo de execução: {temp_mim} segundos")
print(f"Média de tempo de execução: {media_tempo1} segundos")
print(f"Memória máxima consumida: {memori_max} MB")
print(f"Memória mínima consumida: {memori_min} MB")
print(f"Média de memória consumida: {media_memoria1} MB")
```
